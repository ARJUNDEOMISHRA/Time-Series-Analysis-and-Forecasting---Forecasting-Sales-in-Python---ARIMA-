```python
import warnings
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima.model import ARIMA

warnings.filterwarnings("ignore")

# Example p, d, q ranges
p_values = [0, 1, 2, 3]
d_values = [0, 1]
q_values = [0, 1, 2, 3]

# Assume df contains a 'Sales' column
df_train, df_test = df['Sales'][:25], df['Sales'][25:36]

# Initialize best order & RMSE
best_rmse = float("inf")
best_order = None

for p in p_values:
    for d in d_values:
        for q in q_values:
            order = (p,d,q)
            predictions = []
            history = list(df_train)

            try:
                for t in range(len(df_test)):
                    model = ARIMA(history, order=order)
                    model_fit = model.fit()
                    pred_y = model_fit.forecast()[0]
                    predictions.append(pred_y)
                    history.append(df_test.iloc[t])  # update history

                error = np.sqrt(mean_squared_error(df_test, predictions))
                print(f'ARIMA{order} RMSE = {error:.2f}')

                if error < best_rmse:
                    best_rmse = error
                    best_order = order

            except Exception as e:
                print(f"Skipping ARIMA{order} due to error: {e}")
                continue

print(f"\nBest ARIMA order: {best_order} with RMSE = {best_rmse:.2f}")
```
